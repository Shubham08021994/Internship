```python
#getting the url
url = "https://in.via.com/flight/search?returnType=one-way&destination=BOM&bdestination=BOM&destinationL=Mumbai,Mumbai&destinationCity=Mumbai&destinationCN=India&source=DEL&bsource=DEL&sourceL=Delhi,Delhi&sourceCity=Delhi&sourceCN=India&month=1&day=15&year=2023&date=1/15/2023&numAdults=1&numChildren=0&numInfants=0&validation_result=&domesinter=international&livequote=-1&flightClass=ALL&travType=DOM&routingType=ALL&preferredCarrier=&prefCarrier=0&isAjax=false"
driver.get(url)
```

```python
#Airline details
airline_name=[]  
try:
    name=driver.find_elements(By.XPATH,'//div[@class="name js-toolTip"]')
    for i in name:
        airline_name.append(i.text)
except NoSuchElementException:
    airline_name.append('-')
#place
source=[]
try:
    Source=driver.find_elements(By.XPATH,'//div[@class="city"]')
    for i in Source:
        source.append(i.text)
except NoSuchElementException:
    source.append('-')
    
#Source  
Source=[]
for i in range(0,len(source)-1,2):
    Source.append(source[i])
#destination
destination=[]
for i in range(0,len(source)-1,2):
    destination.append(source[i+1])
    
# total stops
stops=[]
try:
    stop=driver.find_elements(By.XPATH,'//span[@class="stops"]')
    for i in stop:
        stops.append(i.text)
except NoSuchElementException:
    stops.append('-')
#departure time 
depature=[]
try:
    time=driver.find_elements(By.XPATH,'//div[@class="time"]')
    for i in time:
        depature.append(i.text)
except NoSuchElementException:
    depature.append('-')
    
#destination
depature_time=[]
arrival_time=[]
for i in range(0,len(depature)-1,2):
    depature_time.append(depature[i])
    arrival_time.append(depature[i+1])
    
    
#duration
duration=[]
try:
    time=driver.find_elements(By.XPATH,'//div[@class="dur"]')
    for i in time:
        duration.append(i.text)
except NoSuchElementException:
    duration.append('-')
    
#fare
price=[]
try:
    fare=driver.find_elements(By.XPATH,'//span[@class="price"]')
    for i in fare:
        price.append(i.text)
except NoSuchElementException:
    price.append('-')
```
